```python
sns.histplot(df1['SAT'])
plt.show()
```

```python
plt.hist(df1['SAT'], color='red')
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
sns.histplot(df1['SAT'], color='red')
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SAT Scores')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rate')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'ro')
plt.xlabel('SAT Scores')
plt.ylabel('Top 10 Percentile')
plt.title('Scatter Plot of SAT Scores vs. Top 10 Percentile')
plt.show()
```

```python
# Create a scatter plot between 'SAT' and 'Top10' columns
plt.plot(df1['SAT'], df1['Top10'], 'ro') # 'ro' specifies red color and circle markers
# Add labels to the x and y axes
plt.xlabel('SAT Scores')
plt.ylabel('Top 10 Percentile')
# Add a title to the plot
plt.title('Scatter Plot of SAT Scores vs. Top 10 Percentile')
# Display the plot
plt.show()
```

```python
sns.scatterplot(x='SAT', y='Top10', data=df1)
plt.xlabel('SAT Scores')
plt.ylabel('Top 10 Percentile')
plt.title('Scatter Plot of SAT Scores vs. Top 10 Percentile')
plt.show()
```
